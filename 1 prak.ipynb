```julia
#1. Написать функцию, вычисляющую НОД двух чисел (или многочленов)

function  gcd_(a::Int,b::Int) 
    while !iszero(b)
        a,b=b,rem(a,b)
    end
    return abs(a)
end

print(gcd_(8,46))
```

```julia
#2. Написать функцию, реализующую расширенный алгоритм Евклида, вычисляющий не только НОД, но и коэффициенты его линейного 
# представления.

function gcdx_(a::Int, b::Int)
    u,v=one(Int),zero(Int)
    u_,v_=v,u
    while !iszero(b)
        k,r=divrem(a,b)
        a,b=b,r #r=a-kb
        u,u_=u_,u-k*u_
        v,v_=v_,v-k*v_
    end
    if (a < 0)
        a,u,v=-a,-u,-v
    end
    return a,u,v
end

print(gcdx_(11,55))
```

```julia
#6,7. Реализовать тип Plolynom{T} (T- тип коэффициентов многочлена) и Обеспечить взаимодействие типов Residue{M} и Polynom{T},
# т.е. добиться, чтобы можно было бы создавать кольцо вычетов многочленов (по заданному модулю) и чтобы можно было создавить
# многочлены с коэффициентами из кольца вычетов.

#RESIDUE
 
struct Rsidue{T,M}
    value::T
    Residue{T,M}(value) where{T,M} = new(value % M)
end

# operator +
Base. +(a::Residue{T,M}, b::Residue{T,M}) where{T,M} = (a.value + b.value) % M
# operator *
Base. *(a::Residue{T,M}, b::Residue{T,M}) where{T,M} = (a.value * b.value) % M
# inverse
function Base.inv(a::Residue{T,M})::Union{Nothing, Residue{T,M}} where{T,M} 
    if gcdex(a.value, M)[1] == 1 
        return Residue{T, M}(gcdex(a.value, M)[2])
    else 
        return nothing
    end
end
# operator /
Base. /(a::Residue{T,M}, b::Residue{T,M}) where{T,M} = a * inv(b)
# un operator -
Base. -(a::Residue{T,M}) where{T,M} = Residue{T,M}(M - a.value)
# operator -
Base. -(a::Residue{T,M}, b::Residue{T,M}) where{T,M} = (a.value + (-b.value)) % M
# operator ==
Base. ==(a::Residue{T,M}, b::Residue{T,M}) where{T,M} = (a.value == b.value) 
# operator <
Base. <(a::Residue{T,M}, b::Residue{T,M}) where{T,M} = (a.value < b.value) 
# operator >
Base. >(a::Residue{T,M}, b::Residue{T,M}) where{T,M} = (a.value > b.value) 
# operator <=
Base. <=(a::Residue{T,M}, b::Residue{T,M}) where{T,M} = (a.value <= b.value) 
# operator >=
Base. >=(a::Residue{T,M}, b::Residue{T,M}) where{T,M} = (a.value >= b.value) 

function gcdex(a::Int, b::Int)
    u,v=one(Int),zero(Int)
    u_,v_=v,u
    while !iszero(b)
        k,r=divrem(a,b)
        a,b=b,r #r=a-kb
        u,u_=u_,u-k*u_
        v,v_=v_,v-k*v_
    end
    if isnegative(a)
        a,u,v=-a,-u,-v
    end
    return a,u,v
end


#POLYNOMIAL

struct Polynomial{T}

    coeff::Vector{T}

    function Polynomial{T}(coeff) where T 
        n = 0
        for c in reverse(coeff)
            if c == 0
                n+=1
            end
        end
        new(coeff[1:end-n])
    end

end

deg(p::Polynomial) = length(p.coeff) - 1

# sum 
function Base. +(p::Polynomial{T}, q::Polynomial{T})::Polynomial{T} where T
    np, nq = length(p.coeff), length(q.coeff)
    if  np >= nq 
        coeff = similar(p.coeff)
        coeff[1:nq] .= (@view(p.coeff[1:nq]) .+ q.coeff) 
    else
        coeff = similar(q.coeff)
        coeff[1:np] .= (p.coeff .+ @view(q.coeff[1:np]))
    end
    i, n = lastindex(coeff), 0
    while i > 0 && coeff[i] == 0
        n += 1
        i -= 1
    end
    resize!(coeff, length(coeff)-n)
    return Polynomial{T}(coeff)
end

# subtraction
function Base. -(p::Polynomial{T}, q::Polynomial{T})::Polynomial{T} where T
    np, nq = length(p.coeff), length(q.coeff)
    if  np >= nq 
        coeff = similar(p.coeff)
        coeff[1:nq] .= (@view(p.coeff[1:nq]) .- q.coeff) 
    else
        coeff = similar(q.coeff)
        coeff[1:np] .= (p.coeff .- @view(q.coeff[1:np]))
    end
    i, n = lastindex(coeff), 0
    while i > 0 && coeff[i] == 0
        n += 1
        i -= 1
    end
    resize!(coeff, length(coeff)-n)
    return Polynomial{T}(coeff)
end

# mult 
function Base. *(p::Polynomial{T}, q::Polynomial{T})::Polynomial{T} where T
    coeff = zeros(T, deg(p) + deg(q)+1)
    for i in eachindex(p.coeff), j in eachindex(q.coeff)
            coeff[i+j - 1] += p.coeff[i]*q.coeff[j]
    end
    i, n = lastindex(coeff), 0
    while i > 0 && coeff[i] == 0
        n += 1
        i -= 1
    end
    resize!(coeff, length(coeff)-n)
    return Polynomial{T}(coeff)
end

# divrem
function Base.divrem(A::Polynomial{T}, B::Polynomial{T}) where T
    B = copy(B.coeff)
    A = copy(A.coeff)
    D = Vector{T}(undef, length(A) - length(B) + 1)
    R = Vector{T}(undef, length(B)-1)
    for i in 1:length(A) - length(B) + 1
        D[i] = A[i] / B[1]
        A[i:end] .-= (D[i] .*B)
    end
    i = findfirst(D .!= 0) 
    D = D[i:end]
    i = findfirst(D .!= 0)
    R = A[i:end]
    return Polynomial{T}(D), Polynomial{T}(R)
end

# divrem
Base.div(p::Polynomial{T}, q::Polynomial{T}) where T = divrem(p,q)[1]
Base.rem(p::Polynomial{T}, q::Polynomial{T}) where T = divrem(p,q)[2]

# sum (int)
Base. +(p::Polynomial{T}, c::T) where T = +(p, Polynomial{T}([c]))
Base. +(c::T, p::Polynomial{T}) where T = +(Polynomial{T}([c]), p)

# subtraction (int)
Base. -(p::Polynomial{T}, c::T) where T = -(p, Polynomial{T}([c]))
Base. -(c::T, p::Polynomial{T}) where T = -(Polynomial{T}([c]), p)

# mult (int)
Base. *(p::Polynomial{T}, c::T) where T = *(p, Polynomial{T}([c]))
Base. *(c::T, p::Polynomial{T}) where T = *(Polynomial{T}([c]), p)

A = Residue{Int,5}(4)
B = Residue{Int,5}(3)
C = A + B
D = A * B
E = -A
a = Polynomial{Int}([1,2,3])
b = Polynomial{Int}([4,5,6])
c = a + b
d = a - b
e = a * b

println(c.coeff)
println(d.coeff)
println(e.coeff)

x = a + 2
y = 2 - a
z = a * 2
println(x.coeff)
println(y.coeff)
println(z.coeff)
```
